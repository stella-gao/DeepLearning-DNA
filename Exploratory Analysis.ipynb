## CpG Islands Across Different Promoter Types

Here, we seek to determine whether the canonical patterns for different promoter types appear in our dataset. Previous findings have suggested that promoters can be categorized into 4 distinct types, three of which differ from each other by the prevalence of CpG islands and the remaining type possessing TCT motifs. We want to verify if these patterns hold in the promoters that we've extracted from public databases.  

We first focus on examples of GO terms that are associated with the types of genes influenced by the different promoter types. Here are the GO terms that we choose to represent each promoter type.  
- Type I: skeletal systems development  
- Type II: translation (or ribosome biogenesis), cell cycle  
- Type III: multicellular organism development, cell differentiation
- Type TCT: translation (or ribosome biogenesis)

### Comparison of Mammalian and Fungi Promoters (Human vs. S. cerevisiae)  

In mammalian promoters, studies have demonstrated evidence for the presence of a considerable quantity of CpG islands. We saw this occurrence ourselves when trying to gather promoter sequences for mice and humans. When using masked sequences from FASTA files available online, we saw that a sizable proportion of promoters for mice and humans were removed from consideration due to the large number of repeats found in these regions. We also know, from previous work, that the presence of CpG islands is not as prevalent in other categories of species. Here, we simply test to see if these observations hold in our data.  

```python
species_list = ['Human','sCer','Mouse']
species_names = ['sapien','cerevisiae','musculus']
ontology_terms = ['cell_cycleP','stressP','translation']
pattern_list = [r'CG','CGCG']

pattern_dict_list = [detectPattern(species_list[i],species_names[i],ontology_terms,pattern_list) for i in range(len(species_list))]
```

```python
%matplotlib inline

f, plot_list = plt.subplots(1,3)

for i in range(3):
    for j in range(len(species_list)):
        plot_list[i].hist(pattern_dict_list[j][pattern_list[0]][ontology_terms[i]],alpha=0.6,normed=True,label=species_list[j])
    plot_list[i].set_title('GO Term: ' + ontology_terms[i] + ', Pattern: ' + pattern_list[0])
    plot_list[i].legend()
f.subplots_adjust(right=2)
```
